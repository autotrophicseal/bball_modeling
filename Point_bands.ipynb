## Optimization

Took me a while to figure out what I was doing, but finally realized I could just try and simulate this in python rather than try to work out a minimization problem with math on paper.

Used ChatGPT to help me write the code so this could be off. Not super familiar with scipy.optimize

```python
payouts = values
probabilities = norm_probabilities
```

```python
def simulation(invested, payouts, probabilities, N=1000):
    gain_loss = [None] * N

    for i in range(N):
        # Choose who wins based on the following
        winner = random.choices(range(len(invested)), probabilities)[0]
        # We add one because we get the payout back + original investment
        gain = (payouts[winner]+1) * invested[winner]
        # We lose all the other money we invested
        loss = sum(invested) - invested[winner]
        gain_loss[i] = gain - loss
    # Return the negative of the gain so we can minimize the function
    return np.mean(gain_loss)
```

```python
initial_invested = [0.5, 0.5, 0.5, 1.5, 3, 1.5, 0.5, 0.5, 0.5]
invested_1 = [0,0,0,3,3,3,0,0,0]
```

```python
from pyomo.environ import *
import random
import numpy as np

# Create a Pyomo model
model = ConcreteModel()

# Define the invested variables
model.invested = Var(range(len(initial_invested)), domain=NonNegativeReals)

# Define the objective function
model.objective = Objective(expr=-simulation([model.invested[i] for i in range(len(initial_invested))], payouts, probabilities), sense=minimize)

# Define the constraint for the sum of invested
model.sum_constraint = Constraint(expr=sum(model.invested[i] for i in range(len(initial_invested))) == 9)

# Solve the optimization problem
SolverFactory('glpk').solve(model)

# Get the optimized 'invested' values
optimized_invested = [model.invested[i].value for i in range(len(initial_invested))]

print("Optimized 'invested' values:", optimized_invested)
```

```python
one = []
two = []
three = []

for i in range(100):
    one.append(simulation(invested_1, payouts, probabilities))
    two.append(simulation(initial_invested, payouts, probabilities))
    three.append(simulation(optimized_invested, payouts, probabilities))
```
