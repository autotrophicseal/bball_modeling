```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

## The Dataset

```python
df = pd.read_csv("force2020_data_unsupervised_learning.csv", index_col='DEPTH_MD')
```

```python
df
```

## Transform the Data

```python
# Remove all rows with na values
# We do it inplace so we don't need to create another dataframe
df.dropna(inplace=True)
```

```python
df
```

```python
# We see that the ranges are significantly different
# We are going to want to standardize
df.describe()
```

```python
scaler = StandardScaler()
```

```python
df[['RHOB_T', 'NPHI_T', 'GR_T', 'PREF_T', 'DTC_T']] = scaler.fit_transform(df[['RHOB', 'NPHI', 'GR', 'PEF', 'DTC']])
```

```python
df
```

```python
def optimise_k_means(data, max_k):
    means = []
    inertias = []
    
    for k in range(1, max_k):
        kmeans = KMeans(n_clusters=k)
        kmeans.fit(data)
        
        means.append(k)
        inertias.append(kmeans.inertia_)
        
    # Generate a plot
    fig = plt.subplots(figsize=(10, 5))
    plt.plot(means, inertias, 'o-')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.grid(True)
    plt.show()
```

```python
optimise_k_means(df[['RHOB_T','NPHI_T']], 10)
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
kmeans.fit(df[['NPHI_T', 'RHOB_T']])
```

```python
df['kmeans_3'] = kmeans.labels_
```

## Plotting the Results

```python
plt.scatter(x = df['NPHI'], y=df['RHOB'], c=df['kmeans_3'])
plt.xlim(-0.1,1)
plt.ylim(3, 1.5)
plt.show()
```
